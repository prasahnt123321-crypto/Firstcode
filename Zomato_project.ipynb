```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/drive/MyDrive/Zomato data .csv")
df.head()

```

```python
df.info()
```

# before the question solveing first data cleaning

# convert the data type of rating

```python
def handlerate(value):
  value = str(value).split("/")
  value = value[0]
  return float(value)
df["rate"] = df["rate"].apply(handlerate)
print(df)



```

#look now the data type of rate is float

# no null values its mean not data missing

```python
df.info()
print(df["rate"].dtype)
```

#  question 1. Type of resturant

```python
ax = sns.countplot(x = df["listed_in(type)"])
plt.xlabel("Type of Resturant" ,c = "green" , size = "20")
plt.ylabel("Count of Resturant",c = "green",size = 20)
for bars in ax.containers:
  ax.bar_label(bars)




```

# mesority of the resturant falls in dinning category.....

# question 2. How many votes has each type of restaurant received from costumers?

```python
gd = df.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({"votes" : gd})
plt.plot(result ,c = "red" ,marker = "o")
plt.xlabel("Type of Resturant" ,c = "green" , size = "20")
plt.ylabel("Total votes",c = "green",size = 20)
```

# conclusion dinning resturants has recived maximum votes

# questin 3. what are the ratings that mejority of restaurant have recived?

```python
plt.hist(df["rate"],bins = 5)
plt.title("rating distribution")
plt.show()
```

# Zomato has observed that most couples order most of their food online.What is their average spending on each order?

```python
couple_data = df["approx_cost(for two people)"]
sns.countplot(x = couple_data)
plt.xlabel("approx_cost(for two people)" ,c = "green" , size = "20")
plt.ylabel("Count",c = "green",size = 20)
```

# question 5 . which mode(ofline or online) has recived the maxminum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = "online_order",y = "rate",data = df)
plt.xlabel("online_order" ,c = "green" , size = "20")
plt.ylabel("ratings",c = "green",size = 20)
```

# conclusion ofline order recived lower rating as campare to online order
